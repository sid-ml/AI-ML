# Lending Club Case Study

## Employment Length Analysis

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
sns.set(color_codes=True)

loan_df = pd.read_csv('loan.csv', low_memory=False)
print("Initial Shape of DF :", loan_df.shape)

# drop columns which containes more than 30% of NA's
loan_df = loan_df.dropna(axis = 1, thresh=int(0.3*len(loan_df)))
print("After Dropping NA columns : ", loan_df.shape)

# drop row which containes emp_length as NA's
loan_df = loan_df[loan_df['emp_length'].notnull()]
print("After Dropping emp_length NA rows : ", loan_df.shape)

empLengthOrder = ['< 1 year','1 year','2 years','3 years','4 years', '5 years','6 years', '7 years', '8 years','9 years','10+ years']

tmpLoanDF = loan_df[['emp_length', 'int_rate', 'loan_amnt', 'term', 'delinq_2yrs', 'dti', 'loan_status']]

fig, ax = plt.subplots(3,1, figsize=(18,18))

# Plot showing Number of Total loan_df as per Employment Length, with status of the loan_df.
s0 = sns.countplot(x="emp_length", hue='loan_status', data=tmpLoanDF, order=empLengthOrder, ax=ax[0])
s0.set(xlabel='Employment Length', ylabel='Number of loan_df')

# Plot showing the range of loan amount given as per employment length
s1 = sns.violinplot(x="emp_length", y="loan_amnt", data=tmpLoanDF, order=empLengthOrder, ax=ax[1], estimator=sum)
s1.set(xlabel='Employment Length', ylabel='Loan Amount')

# Plot showinf number of 30+ days past-due deliquency in last 2 years.
s2 = sns.boxplot(x="emp_length", y="delinq_2yrs",data=tmpLoanDF, palette="Set2", order=empLengthOrder ,ax=ax[2])
s2.set(xlabel='Employment Length', ylabel='delinq_2yrs')

plt.show()

```
